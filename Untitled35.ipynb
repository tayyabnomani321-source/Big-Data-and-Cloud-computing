```python
import matplotlib.pyplot as plt
import networkx as nx

# Create directed graph
G = nx.DiGraph()

# Define nodes
nodes = [
    "Data Sources\n(POS, Online, Inventory)",
    "Data Ingestion\n(Batch & Stream, Kafka/Spark)",
    "Data Storage\n(HDFS / S3 / Delta Lake)",
    "Data Processing\n(PySpark ETL)",
    "Analytics & ML\n(Trend, Forecast, Clustering)",
    "Dashboards / BI\n(Tableau, Power BI, SparkSQL)"
]

# Add nodes
for node in nodes:
    G.add_node(node)

# Define edges (workflow)
edges = [
    (nodes[0], nodes[1]),
    (nodes[1], nodes[2]),
    (nodes[2], nodes[3]),
    (nodes[3], nodes[4]),
    (nodes[4], nodes[5])
]

# Add edges
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(12, 6))
pos = nx.spring_layout(G, seed=42)  # layout

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color="#87CEEB", alpha=0.9)
nx.draw_networkx_edges(G, pos, arrows=True, arrowstyle='-|>', arrowsize=20)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight="bold")

plt.title("RetailChain Big Data Analytics System Architecture", fontsize=14, fontweight="bold")
plt.axis("off")
plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import rand, when, expr
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType

# Initialize Spark
spark = SparkSession.builder.appName("RetailChainDemo").getOrCreate()

# Number of records
num_records = 1000

# Create a DataFrame with just TransactionID
df = spark.range(1, num_records + 1).withColumnRenamed("id", "TransactionID")

# Add Store column using Spark's rand()
df = df.withColumn("Store", when(rand() < 0.33, "Store A")
                   .when(rand() < 0.66, "Store B")
                   .otherwise("Store C"))

# Add Product column
df = df.withColumn("Product", when(rand() < 0.25, "Laptop")
                   .when(rand() < 0.5, "Smartphone")
                   .when(rand() < 0.75, "Headphones")
                   .otherwise("Monitor"))

# Add Amount column (random float between 50 and 2000)
df = df.withColumn("Amount", (rand() * 1950 + 50).cast(FloatType()))

df.show(5)
```

```python
# Aggregate total sales by product
top_products = df.groupBy("Product").sum("Amount").orderBy(col("sum(Amount)").desc())
top_products.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to pandas for plotting
pdf = top_products.toPandas()
plt.figure(figsize=(8,5))
plt.bar(pdf['Product'], pdf['sum(Amount)'], color='skyblue')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales ($)")
plt.show()
```
